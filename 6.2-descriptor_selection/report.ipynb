## Поэтапное изложение процедуры

Данная лабораторная работа проводилась на заведомо шумовых данных: рандомные числа. 

Сначала была обучена линейная модель на основной выборке. Однако результаты перекрёстной проверки этой модели показали отрицательное значение коэффициента детерминации (R²). При этом, даже при прогнозировании на той же выборке, на которой модель обучалась, R² также оказался отрицательным.

Далее была обучена модель с предварительным отбором дескрипторов. Результаты перекрёстной проверки для этой модели показали положительное значение R². Однако при прогнозировании на выборке, не участвовавшей в обучении, R² снова оказался отрицательным. Это закономерный результат, так как изначально данные для обучения содержали лишь шум.

## Выводы 

Была проведена лабораторная работа с базовым алгоритмом отбора дескрипторов. Обучение модели на одних данных и предсказание на других необходимо для оценки её способности обобщать, то есть правильно работать с новыми, ранее невиданными данными. Если модель хорошо предсказывает только на тех данных, на которых обучалась, это может свидетельствовать о переобучении, когда она запоминает детали выборки вместо выявления общих закономерностей.

Перекрёстная проверка позволяет проверить модель на разных частях данных, разделяя выборку на обучающие и тестовые подмножества. Это помогает оценить её устойчивость и способность давать надёжные результаты, независимо от конкретного набора обучающих данных. Такой подход особенно важен для предотвращения переобучения и выбора оптимальных параметров модели.
