## Поэтапное изложение процедуры

Сначала была создана локальная база данных с химическими реакциями в формате RDF. Далее была изучена библиотека через создание примеров ее использования. В ходе выполнения работы были проведены следующие действия:

    - Чтение и запись реакций
    - Изучение реакции как объекта
    - Изучение Атом-Атомное Отображение (ААО)
    - Представление реакции как конденсированный граф реакции
    - Нахождение реакционных центров
