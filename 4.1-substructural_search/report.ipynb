## Поэтапное изложение процедуры

Были сгенерированы дескрипторы в виде отпечатков Моргана. Отпечатки Моргана представляют из себя битовый вектор. Битовый вектор использовался для вычисления индекса Танимото ии индекса косинуса, рассчитанных с помощью библиотеки RDKit. 

Далее битовый вектор был преобразован в численные дескрипторы для вычисления манхэттэнского расстояния, евклидово расстояния, индекса Жаккарда и индекса Дайса из библиотеки scikit-learn. Были построены тепловые карты схожести для первых 10 молекул для более легкой визуализации и интропретации результатов.
