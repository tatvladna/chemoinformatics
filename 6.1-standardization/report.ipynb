## Поэтапное изложение процедуры

В соотвествии с базовым протоколом стандартизации химических данных были стандартизированы локальные данные. Стандартизация проводилась с помощью библиотеки RDKit. Было решено использовать индивидуальные методы стандартизации по очереди. Были применены следующие методы стандартизации:

- Удаление солей
- Нейтрализация зарядов
- Приведение к родительскому фрагменту
- Приведение к стандартной таутомерной форме 
- Удаление стереохимии 
- Сброс изотопов
- Канонизация SMILES
- Добавление и удаление гидрогенов
- Исправление металлов.

Стандартизированные данные были сравнены с исходными.

## Выводы 

В данной работе использовался базовый протокол стандартизации химических данных. Но это не означает, что он может быть применим абсолютно ко всем данным. Важно знать к каким данным он будет примененм и будущее этих данных. И только тогда можно будет сделать подходящий протокол стандартизации.
