## Задачи лабораторной работы

1. Изучение основных возможностей библиотеки **RDkit** и создание простого примера использования

3. [Официальная документация CGRtools](https://github.com/cimm-kzn/CGRtools)

## Поэтапное изложение процедуры

Сначала была собрана локальная база данных химических соединений благодаря веб-ресурсу PubChem. Был собран текстовый файл, содержащий SMILES (линейное представление молекул) некоторых молекул. Далее были загружены некоторые молекулы в MOL формате. Также был собран файл формата SDF c несколькими случайными молекулами.

Далее изучались и сравнивались две популярные хемоинформатические библиотеки через пример их использования. Изучение и сравнение проходило через следующие возможные нужды:

    - Чтение и запись одной молекулы
    - Чтение набора молекул из файлы формата SDF
    - Чтение молекулы по SMILES
    - Чтение молекулы по InChI
    - Запись молекул в SDF 
    - Получение MOL из SMILES

Также была изучена возможность совместного использования двух библиотек.


-----

## Результаты работы (+ таблицы, графики, иллюстрации?, если требуются)

Результаты работы показали, что **RDkit** обладает большим количеством возможностей и способностей для работы с молекулами, включая чтение и запись молекул в различные форматы. **CGRtools** предоставляет более простой интерфейс для работы с молекулами, но его возможности ограничены, чем у **RDkit**.

Изученное совместное использование двух библиотек позволило получить больше информации о молекулах и улучшить работу с ними.

В ходе выполнения лабораторной работы были изучены две популярные хемоинформатические библиотеки на языке Python: **RDKit** и **CGRtools**. библиотека **RDKit** помогает обрабатывать не только сами молекулы, но и дополнительную информацию, которая представлена в файле формата SDF.
